<h3><center>QUESTÃO 1</center></h3>

```python
import pandas as pd
import numpy as np


#Importando via URL cada arquivo sobre o grupo de pesquisa.
grupopesquisa_2019 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=36818', encoding='iso-8859-1')
grupopesquisa_2018 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=33102', encoding='iso-8859-1')

#Importando via URL cada arquivo sobre a relação de pesquisadores.
relacaopesquisadores_2019 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=36821', encoding='iso-8859-1')
relacaopesquisadores_2018 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=33103', encoding='iso-8859-1')

#Importando via URL cada arquivo sobre a relação de pesquisa.
relacaopesquisa_2019 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=36822', encoding='iso-8859-1')
relacaopesquisa_2018 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=33101', encoding='iso-8859-1')

#Importando via URL cada arquivo sobre projeto de extensão.
projetoextensao_2019 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=36482', encoding='iso-8859-1')
projetoextensao_2018 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=33131', encoding='iso-8859-1')

#Importando via URL cada arquivo sobre evento de extensão.
eventoextensao_2019 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=36480', encoding='iso-8859-1')
eventoextensao_2018 = pd.read_csv('https://documento.ifnmg.edu.br/action.php?kt_path_info=ktcore.actions.document.view&fDocumentId=33130', encoding='iso-8859-1')
```

```python
#TRATANDO CADA LINHA DO GRUPO DE PESQUISA, DPOPANDO A 1º LINHA DE FORMA QUE NÃO FIQUE COM REPETIÇÃO.
#FOI FEITO UMA COPIA DE CADA TABELA, PARA NÃO INTERFERIR NA ORIGINAL.
copiagrupopesquisa_2018 = grupopesquisa_2018.drop([0])
copiagrupopesquisa_2018.columns=["NaN","INSTITUIÇÃO","GRUPO","Nome do Coordenador","2º LÍDER","ÁREA PREDOMINANTE"]  #RENOMEANDO AS COLUNAS.
copiagrupopesquisa_2019 = grupopesquisa_2019.drop([0])
copiagrupopesquisa_2019.columns=["INSTITUIÇÃO","GRUPO","Nome do Coordenador","ÁREA PREDOMINANTE","Unnamed: 4","Unnamed: 5"]  #RENOMEANDO AS COLUNAS.

#CONCATENANDO MEU DATAFRAME SOBRE O GRUPO DE PESQUISA
grupopesquisaTOTAL = pd.concat([copiagrupopesquisa_2019,copiagrupopesquisa_2018])
#DROPANDO COLUNAS INDEJESEJADAS
grupopesquisaTOTAL.drop(["Unnamed: 4", "Unnamed: 5", "NaN"], inplace= True, axis=1)
#MOSTRANDO A CONCATENAÇÃO NA TELA
grupopesquisaTOTAL
```

```python
#TRATANDO CADA LINHA DO GRUPO DE RELAÇÃO DE PESQUISADORES, DPOPANDO A 1º LINHA DE FORMA QUE NÃO FIQUE COM REPETIÇÃO.
#FOI FEITO UMA COPIA DE CADA TABELA, PARA NÃO INTERFERIR NA ORIGINAL.
copiarelacaopesquisadores_2018 = relacaopesquisadores_2018.drop([0])
copiarelacaopesquisadores_2018.columns = ["Nome do Coordenador", "Campus"] #RENOMEANDO AS COLUNAS, PARA IGUALAR AS TABELAS.
copiarelacaopesquisadores_2019 = relacaopesquisadores_2019.drop([0])
copiarelacaopesquisadores_2019.columns = ["Nome do Coordenador", "Campus"] #RENOMEANDO AS COLUNAS, PARA IGUALAR AS TABELAS.

#CONCATENANDO MEU DATAFRAME SOBRE O GRUPO DE PESQUISA
relacaopesquisadoresTOTAL = pd.concat([copiarelacaopesquisadores_2019,copiarelacaopesquisadores_2018])
#MOSTRANDO A CONCATENAÇÃO NA TELA
relacaopesquisadoresTOTAL
```

```python
#TRATANDO CADA LINHA DO GRUPO DE RELAÇÃO DE PESQUISA, DPOPANDO A 1º LINHA DE FORMA QUE NÃO FIQUE COM REPETIÇÃO.
#FOI FEITO UMA COPIA DE CADA TABELA, PARA NÃO INTERFERIR NA ORIGINAL.
copiarelacaopesquisa_2018 = relacaopesquisa_2018.drop([0])
copiarelacaopesquisa_2018.columns = ["Campus","Título","Nome do Coordenador"] #RENOMEANDO AS COLUNAS, PARA IGUALAR AS TABELAS.
copiarelacaopesquisa_2019 = relacaopesquisa_2019.drop([0])
copiarelacaopesquisa_2019.columns = ["Campus","Título","Nome do Coordenador","id"] #RENOMEANDO AS COLUNAS, PARA IGUALAR AS TABELAS.

#CONCATENANDO MEU DATAFRAME SOBRE O GRUPO DE PESQUISA
relacaopesquisaTOTAL  = pd.concat([copiarelacaopesquisa_2019,copiarelacaopesquisa_2018])
#DROPANDO COLUNAS INDEJESEJADAS                                   
relacaopesquisaTOTAL.drop(["id"], inplace=True, axis=1)
#MOSTRANDO A CONCATENAÇÃO NA TELA                                   
relacaopesquisaTOTAL
```

```python
#TRATANDO CADA LINHA DO PROJETO DE EXTENSÃO
#FOI FEITO UMA COPIA DE CADA TABELA, PARA NÃO INTERFERIR NA ORIGINAL.
copiaprojetoextensao_2018 = projetoextensao_2018
copiaprojetoextensao_2018.columns = ["Título","Nome do Coordenador","Campus"]#RENOMEANDO AS COLUNAS, PARA IGUALAR AS TABELAS.
copiaprojetoextensao_2019 = projetoextensao_2019
copiaprojetoextensao_2019.columns = ["Título","Nome do Coordenador","Campus","Período de Execução"]#RENOMEANDO AS COLUNAS, PARA IGUALAR AS TABELAS.

#CONCATENANDO MEU DATAFRAME SOBRE O PROJETO DE EXTENSÃO
projetoextensaoTOTAL = pd.concat([copiaprojetoextensao_2019,copiaprojetoextensao_2018])
#DROPANDO COLUNAS INDEJESEJADAS
projetoextensaoTOTAL.drop(["Período de Execução"], inplace=True, axis=1)
#MOSTRANDO A CONCATENAÇÃO NA TELA
projetoextensaoTOTAL
```

```python
#TRATANDO CADA LINHA DO EVENTO DE EXTENSÃO, DPOPANDO A 1º LINHA DE FORMA QUE NÃO FIQUE COM REPETIÇÃO.
copiaeventoextensao_2018 = eventoextensao_2018.drop([0])
copiaeventoextensao_2018.columns = ["CONGRESSO","NaN"] #RENOMEANDO AS COLUNAS, PARA IGUALAR AS TABELAS.
#TRATANDO CADA LINHA DO GRUPO DE EVENTO DE EXTENSÃO
copiaeventoextensao_2019 = eventoextensao_2019
copiaeventoextensao_2019.columns = ["Tipo de Evento","Descrição Sumária do Evento","Campus"]#RENOMEANDO AS COLUNAS, PARA IGUALAR AS TABELAS.

#CONCATENANDO MEU DATAFRAME SOBRE O EVENTO DE EXTENSÃO
eventoextensaoTOTAL = pd.concat([copiaeventoextensao_2019,copiaeventoextensao_2018])
#DROPANDO COLUNAS INDEJESEJADAS
eventoextensaoTOTAL.drop(["NaN","CONGRESSO"], inplace=True, axis=1)
#MOSTRANDO A CONCATENAÇÃO NA TELA
eventoextensaoTOTAL
```

```python
"""AQUI FOI FEITO UMA "CONCATENAÇÃO PARCIAL", OU SEJA, JUNTEI MAIS DA METADE DAS TABELAS COM EXCEÇÃO DA PRIMEIRA,
POIS ESTAVA CAUSANDO UM ERRO QUE ERA O FATO DE TODA TABELA DE GRUPO DE PESQUISADORES VIM COM NaN."""
concatPARCIAL = pd.concat([relacaopesquisadoresTOTAL,relacaopesquisaTOTAL,projetoextensaoTOTAL,eventoextensaoTOTAL])
#AQUI EU DROPEI TODOS OS VALORES NULOS DO DATAFRAME.
concatPARCIAL.dropna(how="all")
```

```python
#AQUI EU PEGUEI A CONCATENAÇÃO PARCIAL FEITA ANTERIORMENTE E FIZ UM MERGE COM A 1° CONCATENAÇÃO QUE TIHA FICADO DE FORA
concatGERAL = grupopesquisaTOTAL.merge(concatPARCIAL,on="Nome do Coordenador")

#DROPEI COLUNAS INDEJESADAS, QUE ATRAPLAHAVA NA ESTETICA DO DATAFRAME FINAL
concatGERAL.drop(["Tipo de Evento","Descrição Sumária do Evento"], inplace=True, axis=1)

#DEFININDO O ÍNDICE PARA SE TORNAR A COLUNA CAMPUS 
concatGERAL.set_index("Campus",inplace=True)

#UM SORT PARA OS VALORES NULOS IREM PRO FINAL.
concatGERAL.sort_values(["Campus"],inplace=True)

#MOSTRANDO A CONCATENAÇÃO GERAL NA TELA
concatGERAL.head()
```

```python
#SALVANDO A CONCATENAÇÃO GERAL(QUESTÃO 01) COMO CSV. 
concatGERAL.to_csv(r"C:\Users\flavi\OneDrive\Área de Trabalho\concatGeral.csv", sep=",",encoding="UTF-8")
```
