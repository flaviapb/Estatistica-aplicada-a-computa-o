<h3> <center>1. Liste as diferenças entre uma lista, um dicionário, um array numpy e uma series pandas?</center></h3>


<h4><center>Respostas</center></h4>

<p>Listas: São compostas por uma ordem de elementos organizados de forma linear, e cada elemento pode ser acessado pelo seu índice, que representa sua posição(Inicia em zero). São ordenadas e delimitadas é feita por colchetes.<br>
    <br>
    Dicionários: São estruturados com dois parâmetros(chaves e valor), servem para implementar mapeamento. E diferente da lista, ela é indexada por chaves e não por índices, e sua delimitação é feita por chaves. <br>
    <br>
    Array Numpy: Podemos definir como uma tabela multidimensional com elementos do mesmo tipo, o que já diferencia das outras estruturas citadas, pois só recebe elementos do mesmo tipo, tem capacidade de vetorização que facilita bastante em qualquer operação, tornando tudo a operação computacionamente mais eficiente.<br>
    <br>
    Series pandas: Uma estrutura de dados, compostas por matrizes unidimensionais capazes de armazenar dados, sua finalidade é criar uma tabela de dados, podendo ser feita, por exemplo, com a junção de listas. Podemos associar, para melhor etendimento, que uma serie representa uma coluna do excel.</p>
