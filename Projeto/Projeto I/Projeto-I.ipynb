<p style="background-color:#D3D3D3;"><strong>Resposta: Sim, com a função compression, estamos "ensinando" ao pandas que queremos descompactar o arquivo, que está sendo importado do site oficial.</strong></p>

<h1><center> <br>(C) Quais as cinco cidades brasileiras onde mais ocorreram acidentes em rodovias federais?</center></h1>


```python
print("\nCinco cidades brasileiras que mais ocorreram acidentes em 2020")
municipios_2020 = dados_2020['municipio'].value_counts().to_frame()
municipios_2020.columns=(["MUNICIPIOS"])
municipios_2020.head()
```

```python
print("\nCinco cidades brasileiras que mais ocorreram acidentes em 2019")
municipios_2019 = dados_2019['municipio'].value_counts().to_frame()
municipios_2019.columns=(["MUNICIPIOS"])
municipios_2019.head()
```

```python
print("\nCinco cidades brasileiras que mais ocorreram acidentes em 2018")
municipios_2018 = dados_2018['municipio'].value_counts().to_frame()
municipios_2018.columns=(["MUNICIPIOS"])
municipios_2018.head()
```

<p  style="background-color:#D3D3D3;"><strong>EXPLICAÇÃO: Primeiramente filtrei os dados da coluna 'município', com a função "value_counts" fiz a contagem dos dados que mais se repetiram, com a ajuda da função “to_frame()” transformei em um DataFrame , depois renomeei a coluna, logo depois utilizei a função 'head()' que vai mostrar as 5 cidades com mais ocorrências de acidentes.</strong></p>

<h1><center> <br>(D) Quantos acidentes com feridos graves aconteceram na Paraíba em 2019?</center></h1>

```python
feridos_2019 = dados_2019['feridos_graves'] >= 1

feridos_2019 = dados_2019[feridos_2019]

feridosPB_2019 = feridos_2019['uf'] == "PB"

totalferidosPB_2019 = feridos_2019[feridosPB_2019]

print("\nQuantidade de acidentes com feridos graves na PB em 2019: {}.".format(totalferidosPB_2019.shape[0]))
```

<p  style="background-color:#D3D3D3;"><strong>EXPLICAÇÃO: A príncipio, filtrei a coluna 'feridos_graves' e armazenei dentro de uma variável (feridos_2019) apenas os dados filtrados que possuem valores maiores ou iguais a 1.<br>
    <br>
    A mesma variável (feridos_2019) vai receber o dataframe (dados_2019), filtrando agora o que estava armazenado na variável (feridos_2019), ou seja, valores maiores ou iguais a 1. Logo depois, foi necessário a criação de uma nova variável (feridosPB_2019) para filtrar agora a coluna 'uf' que vai verificar se o estado é PB. Por último, criei a ultima variável (totalferidosPB_2019) para receber os dados fltrados pelas duas variáveis citadas anteriormente, para finalizar, printei na tela utilizando a função shape[0].</strong></p>

<h1><center> <br>TAREFA 3 </center></h1>

<h1><center> <br>(A) Ranking de acidentes por estado para todos os anos (um só arquivo combinando a informação dos três anos);</center></h1>



```python
print("\n---------------RANKING DO MENOR PARA O MAIOR-------------------")

acidente_2020 = dados_2020['uf'].value_counts().reset_index()
acidente_2019 = dados_2019['uf'].value_counts().reset_index()
acidente_2018 = dados_2018['uf'].value_counts().reset_index()

acidente_UF = pd.concat([acidente_2020,acidente_2019,acidente_2018], keys=["2020","2019","2018"] ,axis=1)
acidente_UF.rename(columns=({"index":"ESTADOS","uf":"ACIDENTES"}),inplace=True)
acidente_UF.index.name = "COLOCAÇÃO"
acidente_UF.reindex(index=acidente_UF.index[::-1])
```

<p style="background-color:#D3D3D3;"><strong>EXPLICAÇÃO: Para contar a quantidade de acidentes, criei 3 variáveis (acidente_2020, acidente_2019, acidente_2018) que armazenam a filtagrem de dados da coluna 'uf', utilizei "value_counts" para fazer a contagem dos dados e "reset_index()" para resetar os índices de cada coluna. Em seguida, juntei os dados com concat, renomeei as colunas, adicionei um nome para o índices, e por fim mudei a ordem do DataFrame para crescente.</strong></p>

```python
acidente_UF.to_csv("totalAcidentesUF.csv",index=True)

#Caso, deseje executar, remova as "#":

#a = pd.read_csv("totalAcidentesUF.csv")
#a
```

<h1><center> <br>TAREFA 4 </center></h1>

<h1><center> <br>(A) Qual a causa mais frequente e a mais rara de acidentes registrados nos três anos?</center><h1>

```python
frequencia_2020 = dados_2020['causa_acidente']
frequencia_2019 = dados_2019['causa_acidente']
frequencia_2018 = dados_2018['causa_acidente']

totalFrquencias = pd.concat([frequencia_2020.to_frame(),frequencia_2019.to_frame(),frequencia_2018.to_frame()])
frequencia_comum = totalFrquencias.value_counts().to_frame().head(1)
frequencia_rara = totalFrquencias.value_counts().to_frame().tail(12)

final = pd.concat([frequencia_comum,frequencia_rara])
final.reset_index(inplace=True)
final.columns = ["ACIDENTE",'VALOR']
final.index = ["FREQUENCIA COMUM","FREQUENCIA RARA","FREQUENCIA RARA","FREQUENCIA RARA","FREQUENCIA RARA","FREQUENCIA RARA","FREQUENCIA RARA","FREQUENCIA RARA","FREQUENCIA RARA","FREQUENCIA RARA","FREQUENCIA RARA","FREQUENCIA RARA","FREQUENCIA RARA"]
final
```

<p style="background-color:#D3D3D3;"><strong>EXPLICAÇÃO: Utilizei "head(1)" para obter a frequência mais comum e usei "tail(12)" para obter a frequência mais rara. Visto que "tail()" tem a função de retornar as últimas n linhas (5 é o número padrão de elementos), utilizei n = 12, visto que as causas mais raras estavam nas últimas 12 linhas.</strong></p>

<h1><center> <br>(B) Qual a proporção de pessoas ilesas e de feridos graves por mês em cada um dos anos considerados?
            (proporção =  numero de ilesos ou feridos graves/numero de pessoas envolvidas no acidente);</center></h1>


```python
ilesos2020 = dados_2020.groupby(pd.DatetimeIndex(dados_2020.data_inversa).month)['ilesos'].sum()
feridos2020 = dados_2020.groupby(pd.DatetimeIndex(dados_2020.data_inversa).month)['feridos_graves'].sum()
pessoas2020 = dados_2020.groupby(pd.DatetimeIndex(dados_2020.data_inversa).month)['pessoas'].sum()
proporcaoilesos2020 = ilesos2020/pessoas2020
proporcaoferidos2020 = feridos2020/pessoas2020

total2020 = pd.concat([proporcaoilesos2020,proporcaoferidos2020], axis=1)
total2020.index = ['JANEIRO', 'FEVEREIRO', 'MARÇO', 'ABRIL', 'MAIO', 'JUNHO', 'JULHO', 'AGOSTO', 'SETEMBRO', 'OUTUBRO', 'NOVEMBRO', 'DEZEMBRO']
total2020.columns= ["ILESOS","FERIDOS"]
total2020
```

```python
ilesos2019 = dados_2019.groupby(pd.DatetimeIndex(dados_2019.data_inversa).month)['ilesos'].sum()
feridos2019 = dados_2019.groupby(pd.DatetimeIndex(dados_2019.data_inversa).month)['feridos_graves'].sum()
pessoas2019 = dados_2019.groupby(pd.DatetimeIndex(dados_2019.data_inversa).month)['pessoas'].sum()
proporcaoilesos2019 = ilesos2019/pessoas2019
proporcaoferidos2019 = feridos2019/pessoas2019

total2019 = pd.concat([proporcaoilesos2019,proporcaoferidos2019], axis=1)
total2019.index = ['JANEIRO', 'FEVEREIRO', 'MARÇO', 'ABRIL', 'MAIO', 'JUNHO', 'JULHO', 'AGOSTO', 'SETEMBRO', 'OUTUBRO', 'NOVEMBRO', 'DEZEMBRO']
total2019.columns= ["ILESOS","FERIDOS"]
total2019
```

```python
ilesos2018 =  dados_2018.groupby(pd.DatetimeIndex(dados_2018.data_inversa).month)['ilesos'].sum()
feridos2018 = dados_2018.groupby(pd.DatetimeIndex(dados_2018.data_inversa).month)['feridos_graves'].sum()
pessoas2018 = dados_2018.groupby(pd.DatetimeIndex(dados_2018.data_inversa).month)['pessoas'].sum()
proporcaoilesos2018 = ilesos2018/pessoas2018
proporcaoferidos2018 = feridos2018/pessoas2018

total2018 = pd.concat([proporcaoilesos2019,proporcaoferidos2019], axis=1)
total2018.index = ['JANEIRO', 'FEVEREIRO', 'MARÇO', 'ABRIL', 'MAIO', 'JUNHO', 'JULHO', 'AGOSTO', 'SETEMBRO', 'OUTUBRO', 'NOVEMBRO', 'DEZEMBRO']
total2018.columns= ["ILESOS","FERIDOS"]
total2018
```

<p  style="background-color:#D3D3D3;"><strong>EXPLICAÇÃO: Criei 3 variáveis (ilesos2018, feridos2018, pessoas2018), após isso, com a função "groupby" agrupei, em conjunto da função "DatetimeIndex" que teve como parâmetro o nosso Dataframe especificando qual coluna (dados_2018.data_inversa) e o paramêtro "month", tendo como objetivo retornar qual mês iremos trabalhar. Filtrei as colunas "ilesos","feridos_graves" e "pessoas", após isso, com a função "sum()" somei todos os valores armazenados nas variáveis. A partir desses resultados, criei mais duas variáveis (proporcaoilesos2018, proporcaoferidos2018) que armazenam as proporções de ilesos e feridos. Por fim, com a função "concat" fiz a junção dos dados armazenados anteriormente, renomeei as colunas e os índices.<br>
    <br>
     PS: A explicação acima se aplica para os três anos.</strong></p>

<h1><center> <br>(C) Mostre, reorganizando o conjunto de dados, se a pandemia conseguiu diminuir ou não a incidência de acidentes nas rodovias federais.</center></h1>



```python
media = (dados_2018.shape[0] + dados_2019.shape[0])/2
diferenca = dados_2020.shape[0] - media
total = pd.DataFrame([dados_2018.shape[0],dados_2019.shape[0],dados_2020.shape[0],diferenca])
total.columns = ['TOTAL']
total.index = ["2018","2019","2020","Diferença/média"]
total
```

<p  style="background-color:#D3D3D3;"><strong>EXPLICAÇÃO:  A princípio, fiz a média dos dois anos anteriores 2018 e 2019, para poder calcular a diferença com o ano de 2020 (ano que foi afetado pela pandemia).<br>Criei um DataFrame para guardar o número de acidentes de cada ano com a diferença de antes e depois, variável "diferença", da pandemia.</strong></p>
