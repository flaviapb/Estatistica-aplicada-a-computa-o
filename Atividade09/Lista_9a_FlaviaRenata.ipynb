```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dadosTitanic = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv")
dadosQualitativos = dadosTitanic[[ "Survived", "Pclass","Sex","Embarked"]]
dadosQualitativos.loc[:10].style.set_properties(**{'background-color': 'Lavender',
                           'color': 'grey21'})
```

```python
cores = ("PaleTurquoise","LightPink")
espalhar = (0.1, 0.0)
tipo = ("Masculino","Feminino")
dadosQualitativos["Sex"].value_counts().plot.pie(ylabel=" ",title="SEXO", colors = cores, explode = espalhar, autopct = "%1.1f%%", labels = tipo, textprops = dict(color ="Black"))
plt.show()
```

```python
cores = ("LightSlateGray","LightSkyBlue")
espalhar = (0.1, 0.0)
tipo = ("NÃO SOBREVIVEU","SOBREVIVEU")
dadosQualitativos["Survived"].value_counts().plot.pie(ylabel=" ",title="SOBREVIVENTES", colors = cores, explode = espalhar, autopct = "%1.1f%%", labels = tipo, textprops = dict(color ="Black"))
plt.show()
```

```python
cores = ("Salmon", "LightSalmon", "DarkSalmon")
espalhar = (0.0, 0.1, 0.1)
tipo = ("3° CLASSE","1° CLASSE", "2° CLASSE")
dadosQualitativos["Pclass"].value_counts().plot.pie(ylabel=" ",title="CLASSE DO BILHETE", colors = cores, explode = espalhar, autopct = "%1.1f%%", labels = tipo, textprops = dict(color ="Black"))
plt.show()
```

```python
cores = ("Lavender", "MistyRose", "LavenderBlush")
espalhar = (0.0, 0.1, 0.2)
tipo = ("SOUTHAMPTON","CHERBOURG", "QUEENSTOWN")
dadosQualitativos["Embarked"].value_counts().plot.pie(ylabel=" ",title="PORTO", colors = cores, explode = espalhar, autopct = "%1.1f%%", labels = tipo, textprops = dict(color ="Black"))
plt.show()
```

```python
cores = ("PaleTurquoise","LightPink")
label=["MASCULINO","FEMININO"]
labels = np.arange(len(label))
dadosQualitativos["Sex"].value_counts().plot.bar(color = cores, title = "SEXO")
plt.xticks(labels, label)
plt.show()
```

```python
cores = ("LightSlateGray","LightSkyBlue")
label=["NÃO SOBREVIVEU","SOBREVIVEU"]
labels = np.arange(len(label))
dadosQualitativos["Survived"].value_counts().plot.bar(color = cores, title = "SOBREVIVENTES")
plt.xticks(labels, label)
plt.show()
```

```python
label = ("3° CLASSE","1° CLASSE", "2° CLASSE")
labels = np.arange(len(label))
dadosQualitativos["Pclass"].value_counts().plot.bar(color = "Crimson", title = "CLASSE DO BILHETE")
plt.xticks(labels, label)
plt.show()
```

```python
label = ("SOUTHAMPTON","CHERBOURG", "QUEENSTOWN")
labels = np.arange(len(label))
dadosQualitativos["Embarked"].value_counts().plot.bar(color = "Lavender", title = "PORTO")
plt.xticks(labels, label)
plt.show()
```
