```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'Nome': ['Joao da Silva ', 'Pedro Fernandes' , 'Maria Freitas','Paula Gonçalves','Ana Freitas','Luiz Costa','Andre Souza','Patricia Silva', 'Regina Lima','Alfredo Souza','Margarete Cunha','Pedro Barbosa','Ricardo Alves','Marcio Rezende','Ana Carolina Chaves'],
                   'Idade': [36,51,26,25,29,53,42,38,35,45,26,37,24,31,29],
                    'Salario': [6300,5700,4500,3800,3200,7300,7100,5600,6400,7000,3700,6500,4000,5100,4500]})

df
```

---

<h1><center>PRIMEIRA QUESTÃO</center></h1>

```python
maiorIdade = df['Idade'].max()
menorIdade = df['Idade'].min()
diferenca = maiorIdade - menorIdade
tamanho = len(df)
numerodeclasses = 5
tamanhodaclasse = diferenca/numerodeclasses
dadospratabela = pd.DataFrame({'MAIOR IDADE':[maiorIdade],
                              'MENOR IDADE': [menorIdade],
                              'DIFERENÇA':[diferenca],
                              'TAMANHO':[tamanho],
                              'NÚMERO DE CLASSES':[numerodeclasses],
                              'TAMANHO DA CLASSE':[tamanhodaclasse]})

dadospratabela.round(0)
```

```python
#PREPARANDO OS DADOS PARA TABELA DE PREQUENCIA

dados = pd.DataFrame({'CLASSE 1': ['24 |- 30'],
                     'CLASSE 2': ['30  |- 36'],
                     'CLASSE 3': ['36  |- 42'],
                     'CLASSE 4': ['42  |- 48'],
                     'CLASSE 5': ['48  |- 54']})

dados.index = ['CLASSES']
dados
```

```python
#ANALÍSE DE CADA CLASSE
    #24 |- 30 = 24 25 26 26 29 29 = 6
    #30 |- 36 = 31 35 = 2
    #36 |- 42 = 36 37 38 = 3
    #42 |- 48 = 42 45 = 2
    #48 |- 54 = 51 53 = 2 

totalfr = ((6/15*100))+((2/15*100))+((3/15*100))+((2/15*100))+((2/15*100))

tabelafrequencia = pd.DataFrame({
                      'CLASSE': ['24 |- 30', '30 |- 36', '36 |- 42', '42 |- 48', '48 |- 54', ' ', ' ', ' '],
                      'FREQUÊNCIA \ FI':['6','2','3','2','2',' ',' ','15'],
                      'FREQUÊNCIA RELATIVA \ FR':["{:.2f}".format((6/15*100)),"{:.2f}".format((2/15*100)),"{:.2f}".format((3/15*100)),"{:.2f}".format((2/15*100)),"{:.2f}".format((2/15*100)),' ',' ',"{:.0f}".format(totalfr)],
                      'FREQUÊNCIA ACUMULADA \ FA':['6','8','11','13','15',' ',' ','15']})

tabelafrequencia.index= ['ORDEM 1','ORDEM 2','ORDEM 3','ORDEM 4',' ORDEM 5',' ',' ',' TOTAL DADOS']
tabelafrequencia
```

<h3><center>FINAL DA PRIMEIRA QUESTÃO</center></h3>

```python
#PREPARANDO OS DADOS PARA 2° TABELA DE PREQUENCIA

dados = pd.DataFrame({'CLASSE 1': ['3200 |- 4020'],
                     'CLASSE 2': ['4020  |- 4840'],
                     'CLASSE 3': ['4840  |- 5660'],
                     'CLASSE 4': ['5660  |- 6480'],
                     'CLASSE 5': ['6480  |- 7300']})

dados.index = ['CLASSES']
dados
```
